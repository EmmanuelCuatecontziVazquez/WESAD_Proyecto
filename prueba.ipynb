```python
import pickle
import numpy as np

ruta = 'WESAD/S2/S2.pkl'
with open(ruta, 'rb') as f:
    data = pickle.load(f, encoding='latin1')

label = np.array(data['label'])

# Ver cuándo ocurren las transiciones hacia stress (label=2)
# para entender la estructura temporal
changes = np.where(np.diff(label) != 0)[0]
print('Índices de cambio de estado:')
for idx in changes:
    print(f'  t={idx} ({idx/700:.1f}s): {label[idx]} → {label[idx+1]}')

# Ver duración de cada segmento
print(f'\nDuración total: {len(label)/700:.1f}s = {len(label)/700/60:.1f} min')
```
